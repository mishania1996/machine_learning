```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('data_predicting_disaster/train.csv')
test_df = pd.read_csv('data_predicting_disaster/test.csv')
print(train_df.shape)
print(train_df.head())
print(test_df.shape)
print(test_df.head())
```

Using Logistic Regression model. pipeline chains preprocessing, feature extraction, and model fitting, into a single object.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegressionCV
from sklearn.pipeline import make_pipeline
from sklearn import metrics

train_df['keyword'] = train_df['keyword'].fillna('')

# Define a function to emphasize the keyword
def emphasize_keyword(keyword, text, weight=3):
    return (keyword + ' ') * weight + text

# Apply the function to create the combined column
train_df['combined'] = train_df.apply(lambda row: emphasize_keyword(row['keyword'], row['text']), axis=1)

vec = CountVectorizer()
clf = LogisticRegressionCV(max_iter=1000)
pipe = make_pipeline(vec,clf)
pipe.fit(train_df["combined"],train_df["target"])

y = train_df["target"]
y_pred = pipe.predict(train_df["combined"])
report = metrics.classification_report(y, y_pred)
print(report)
print(metrics.accuracy_score(y,y_pred))
```
